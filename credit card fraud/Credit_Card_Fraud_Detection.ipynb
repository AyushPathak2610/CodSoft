```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning

```python
df_fraud = df[df['Class']==1]
df_genuine = df[df['Class']==0]
```

```python
df_fraud.shape
```

```python
df_genuine.shape
```

```python
df_genuine = df_genuine.sample(n=492)
df_genuine.shape
```

```python
df_sampled = pd.concat([df_genuine,df_fraud],axis=0)
df_sampled.info()
```

# Exploratory Data Analysis

```python
sns.distplot(df_fraud['Time'],label='fraudulent',hist=False);
sns.distplot(df_genuine['Time'],label='non fraudulent',hist=False);
```

```python
sns.distplot(df_fraud['Amount'],label='fraudulent',hist=False);
sns.distplot(df_genuine['Amount'],label='non fraudulent',hist=False);
```

# Data Split and Preparation

```python
from sklearn.model_selection import train_test_split
X = df_sampled.drop(['Class'],axis=1)
y = df_sampled['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
df_sampled.describe()
```

```python
plt.boxplot(df_sampled['Amount'], vert=False)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
X_train['Amount'] = scaler.fit_transform(X_train[['Amount']])
X_test['Amount'] = scaler.transform(X_test[['Amount']])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train['Time'] = scaler.fit_transform(X_train[['Time']])
X_test['Time'] = scaler.transform(X_test[['Time']])
```

```python
time = X_test['Time']
X_test['Time'] = (time - time.min()) / (time.max() - time.min())
```

```python
X_test.describe()
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
lr = LogisticRegression()
param_grid = {'C': [0.01, 0.1, 1, 10]}
grid_search_lr = GridSearchCV(lr,param_grid,cv=5,scoring='roc_auc',verbose=1)
grid_search_lr.fit(X_train,y_train)
grid_search_lr.best_params_
```

```python
lr_model = LogisticRegression(C=0.01)
lr_model.fit(X_train,y_train)
y_train_pred = lr_model.predict(X_train)
y_test_pred = lr_model.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train,y_train_pred))
```

```python
print(classification_report(y_test,y_test_pred))
```

# Neural Network model


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization, Dropout

model_an = Sequential()
```

```python
model_an.add(InputLayer(input_shape=(X_train.shape[1],)))
model_an.add(Dense(units=32, activation='relu'))
model_an.add(BatchNormalization())
model_an.add(Dropout(0.2))
model_an.add(Dense(units=16, activation='relu'))
model_an.add(BatchNormalization())
model_an.add(Dropout(0.2))
model_an.add(Dense(units=1, activation='sigmoid'))
```

```python
model_an.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_an.summary()
```

```python
model_an.fit(X_train, y_train, epochs=100, batch_size=64, validation_split=0.3)
```

```python
y_pred_an = model_an.predict(X_test)
y_pred_an = (y_pred_an > 0.5).astype(np.int32)
print(classification_report(y_test, y_pred_an))
```

```python
model_an2 = Sequential()
model_an2.add(InputLayer(input_shape=(X_train.shape[1],)))
model_an2.add(Dense(units=64, activation='relu'))
model_an2.add(BatchNormalization())
model_an2.add(Dense(units=32, activation='relu'))
model_an2.add(BatchNormalization())
model_an2.add(Dense(units=16, activation='relu'))
model_an2.add(BatchNormalization())
model_an2.add(Dense(units=1, activation='sigmoid'))
model_an2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_an2.summary()
model_an2.fit(X_train, y_train, epochs=100, batch_size=64, validation_split=0.3)
```

```python
y_pred_an2 = model_an.predict(X_test)
y_pred_an2 = (y_pred_an2 > 0.5).astype(np.int32)
print(classification_report(y_test, y_pred_an2))
```

# XGBOOST CLASSIFIER

```python
from xgboost import XGBClassifier
params = {
    'n_estimators': [150, 250, 350],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.001, 0.01, 0.1],
}
model_xgb = XGBClassifier()
grid_xgb = GridSearchCV(model_xgb,param_grid=params,cv=5,scoring='roc_auc',verbose=1)
grid_xgb.fit(X_train,y_train)
grid_xgb.best_params_
```

```python
model_xgb = XGBClassifier(learning_rate=0.1,max_depth=6,n_estimators=350)
model_xgb.fit(X_train,y_train)
y_test_pred_xgb = model_xgb.predict(X_test)
print(classification_report(y_test, y_test_pred_xgb))
```
